## Análisis

```python
df = df[df["recomendaciones_totales"] >= 6]
df_sample1 = df[(df["price_overview"] >= 0.01) & (df["price_overview"] <= 20.00)].sample(n=400, random_state=1)
df_sample2 = df[(df["price_overview"] >= 20.01) & (df["price_overview"] <= 40.00)].sample(n=300, random_state=1)
df_sample3 = df[(df["price_overview"] >= 40.01) & (df["price_overview"] <= 80.01)].sample(n=100, random_state=1)
df_sample = pd.concat([df_sample1, df_sample2, df_sample3], axis=0).reset_index(drop=True)

modelos_nombres = ['v_resnet', 'v_convnext', 'v_clip']
titulos_modelos = ['ResNet', 'ConvNeXt', 'CLIP']

fig = make_subplots(
    rows=2, cols=3,
    subplot_titles=(
        f'PCA - {titulos_modelos[0]}', f'PCA - {titulos_modelos[1]}', f'PCA - {titulos_modelos[2]}',
        f't-SNE - {titulos_modelos[0]}', f't-SNE - {titulos_modelos[1]}', f't-SNE - {titulos_modelos[2]}'
    ),
    horizontal_spacing=0.02,
    vertical_spacing=0.05
)

df_sample['log_recs'] = np.log10(df_sample['recomendaciones_totales'] + 1)

for i, mod in enumerate(modelos_nombres):
    # PCA
    fig.add_trace(
        go.Scatter(
            x=df_sample[f'pca_{mod}_1'],
            y=df_sample[f'pca_{mod}_2'],
            #z=df_sample[f'pca_{mod}_3'],
            mode='markers',
            marker=dict(size=10, color=df_sample['price_overview'], colorscale='Viridis', opacity=0.8),
            text=df_sample['id'],
            hoverinfo='name',
            name=f'PCA {mod}'
        ),
        row=1, col=i+1
    )
    
    # TSNE
    fig.add_trace(
        go.Scatter(
            x=df_sample[f'tsne_{mod}_1'],
            y=df_sample[f'tsne_{mod}_2'],
            #z=df_sample[f'tsne_{mod}_3'], 
            mode='markers',
            marker=dict(size=10, color=df_sample['price_overview'], colorscale='Viridis', opacity=0.8),
            text=df_sample['id'],
            hoverinfo='name',
            name=f't-SNE {mod}'
        ),
        row=2, col=i+1
    )

fig.update_layout(
    title_text='Embeddings de imágenes reducidos a 2 dimensiones donde el color es el precio de los juegos',
    height=900, 
    width=1300,
    showlegend=False
)

fig.update_scenes(xaxis_showticklabels=False, yaxis_showticklabels=False, zaxis_showticklabels=False)

fig.show()
```

Hemos visualizado los vectores representativos de las imágenes generados por tres modelos. Además para poder visualizarlos hemos usado dos técnicas de reducción de dimensionalidad diferentes, PCA (arriba) y t-SNE (abajo). Además el color es el precio de los juegos. En la gráfica intentamos buscar agrupaciones de puntos con un color parecido, lo que podría querer decir que el modelo está agrupando correctamente por precios solamente conociendo las imágenes (al menos de esa agrupación). Parece que CLIP es el modelo que mejor funciona.

```python
matrix_resnet = np.vstack(df_sample['v_resnet'].values)
matrix_convnext = np.vstack(df_sample['v_convnext'].values)
matrix_clip = np.vstack(df_sample['v_clip'].values)

# 2. Función auxiliar actualizada para trabajar con las matrices directamente
def obtener_vecinos_modelo(matrix, query_idx_local):
    nn = NearestNeighbors(n_neighbors=5, metric='cosine')
    nn.fit(matrix)
    _, indices = nn.kneighbors(matrix[query_idx_local].reshape(1, -1))
    return df_sample.iloc[indices[0][1:]][['id', 'price_overview']].values.tolist()

print("COMPARATIVA DE VECINOS POR MODELO (ESPACIO ORIGINAL)\n" + "="*60)

puntos_test = df_sample.sample(10)
for idx_pandas in puntos_test.index:
    idx_local = df_sample.index.get_loc(idx_pandas)
    id_objetivo = df_sample.iloc[idx_local][['id', 'price_overview']].values
    
    # Calculamos vecinos para cada arquitectura
    vecinos_resnet = obtener_vecinos_modelo(matrix_resnet, idx_local)
    vecinos_convnext = obtener_vecinos_modelo(matrix_convnext, idx_local)
    vecinos_clip = obtener_vecinos_modelo(matrix_clip, idx_local)
    
    # Formateo de salida
    sum = 0
    for i in range(2):
        sum += vecinos_resnet[i][1]
        sum += vecinos_convnext[i][1]
        sum += vecinos_clip[i][1]
    print(f"predicción: {sum/6} €")
    print(f"ID Objetivo: {id_objetivo}")
    print(f"  Vecinos (ResNet18 - Formas/Color):   {vecinos_resnet}")
    print(f"  Vecinos (ConvNeXt - Textura/Calidad): {vecinos_convnext}")
    print(f"  Vecinos (CLIP - Semántica/Estilo):    {vecinos_clip}")
    print("-" * 60)
```

En la salida de esta celda se muestra un k-nn manual que hemos usado para probar la relación de los vectores con la variable salida. Hemos intentado predecir el precio únicamente con las imágenes, y aunque no ha salido muy bien, se puede apreciar que a veces si se acerca. Esto no tiene por que ser negativo, ya que no esperábamos hacer un buen modelo tan solo con esta variable, pero confiamos en que junto al resto de variables tendremos un buen resultado.

Todos los pares en la salida de forma [[valor1, valor2]] representan un juego con los valores [[appid, precio]]. Y la predicción que aparece encima de cada juego es una media de los 2 vecinos cercanos de cada modelo. 

# Clustering de Vectores

## HDBSCAN

```python
X = np.vstack(df['v_clip'].values)

reducer = umap.UMAP(n_components=10, random_state=42) 
X_reduced = reducer.fit_transform(X)
X_reduced

clusterer = hdbscan.HDBSCAN(min_cluster_size=8)

labels = clusterer.fit_predict(X_reduced)

df['cluster'] = labels
```

```python
reducer_2d = umap.UMAP(n_neighbors=15, n_components=2, random_state=42)

X_2d = reducer_2d.fit_transform(X)

plt.figure(figsize=(10, 8))
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=labels, cmap='tab20', s=5)
plt.show()
```

```python
df['cluster'].value_counts()
```

Los clusters no están bien agrupados, hay mucho ruido y no es capaz de categorizar muchas de las imágenes

Probamos con menos vecinos:

```python
X = np.vstack(df['v_clip'].values)

reducer = umap.UMAP(n_components=10, random_state=42) 
X_reduced = reducer.fit_transform(X)
X_reduced

clusterer = hdbscan.HDBSCAN(min_cluster_size=3)

labels = clusterer.fit_predict(X_reduced)

df['cluster'] = labels
```

```python
reducer_2d = umap.UMAP(
    n_neighbors=15,
    n_components=2,
    random_state=42
)

X_2d = reducer_2d.fit_transform(X)

plt.figure(figsize=(10, 8))
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=labels, cmap='tab20', s=5)
plt.show()
```

En general HDBSCAN no es capaz de encontrar correctamente los clusters, identifica la mayoría como ruido

## KMEANS

```python
X = np.vstack(df['v_clip'].values)

reducer = umap.UMAP(n_components=10, random_state=42) 
X_reduced = reducer.fit_transform(X)
X_reduced

kmeans = KMeans(n_clusters=20,random_state=42)

labels = kmeans.fit_predict(X_reduced)

df['cluster'] = labels
```

Graficamos:

```python
reducer_2d = umap.UMAP(
    n_neighbors=15,
    n_components=2,
    random_state=42
)

X_2d = reducer_2d.fit_transform(X)

plt.figure(figsize=(10, 8))
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=labels, cmap='tab20', s=5)
plt.show()
```

```python
df['cluster'].value_counts()
```

Con 20 clusters observamos que encuentra los clusters correctamente, pero hay muchos dispersos y no están agrupados correctamente.

Vamos a probar con 10:

```python
X = np.vstack(df['v_clip'].values)

reducer = umap.UMAP(n_components=10, random_state=42) 
X_reduced = reducer.fit_transform(X)
X_reduced

kmeans = KMeans(
    n_clusters=10,
    random_state=42
)

labels = kmeans.fit_predict(X_reduced)

df['cluster'] = labels
```

```python
reducer_2d = umap.UMAP(
    n_neighbors=15,
    n_components=2,
    random_state=42
)

X_2d = reducer_2d.fit_transform(X)

plt.figure(figsize=(10, 8))
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=labels, cmap='tab20', s=5)
plt.show()
```

```python
df['cluster'].value_counts()
```

Con 10 clusters las agrupaciones son más uniformes y se reduce el ruido.

# Reducción de Dimensionalidad

## PCA

Vamos a ver la varianza explicada:

```python
X = np.vstack(df['v_clip'].values)
pca = PCA()
X = pca.fit_transform(X)
cumvar = np.cumsum(pca.explained_variance_ratio_) 

plt.plot(cumvar)
plt.show()
```

Vemos que para un 0.9 de varianza explicada tenemos que reducir a 300 columnas, no es la más óptimo y sigue explotando la dimensionalidad. Además de que CLIP genera embeddings que no tienen relación lineal, por tanto es mejor usar técnicas de reducción no lineales.

## TSNE

Vamos a ver como se comporta en clustering:

```python
X = np.vstack(df['v_clip'].values)

reducer = TSNE(
    n_components=2,       
    random_state=42
)
X_reduced = reducer.fit_transform(X)

kmeans = KMeans(n_clusters=20,random_state=42)
labels = kmeans.fit_predict(X_reduced)
df['cluster'] = labels

plt.figure(figsize=(10, 8))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=labels, cmap='tab20', s=5)
plt.show()
```

```python
df['cluster'].value_counts()
```

TSNE se comporta mejor en los clusters y crea fronteras más definidas para los clusters. A diferencia de UMAP que trata de preservar la estructura global del dataset, T-SNE se centra en preservar la estructura local de cada punto, sus vecinos cercanos.
