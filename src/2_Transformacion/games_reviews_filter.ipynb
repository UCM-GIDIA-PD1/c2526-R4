```python
import gzip
import json
import pandas as pd
```

```python
input_file = "games_info.jsonl.gz"

with gzip.open(input_file, "rt", encoding="utf-8") as f:
        data = [json.loads(line) for line in f if line.strip()]
    
df = pd.DataFrame(data)
```

```python
def get_total_reviews(x):
    if not isinstance(x,dict):
        return []
    if x.get("rollups"):
        return x.get("rollups").get("total_recommendations")
def get_name(x):
    if isinstance(x,dict):
        return x.get("name")
    else:
        return None
```

```python
df["name"] = df["appdetails"].apply(lambda x : get_name(x))
df["total_reviews"] = df["appreviewhistogram"].apply(lambda x : get_total_reviews(x))

df = df[["name","total_reviews"]]
```

```python
df_sorted = df.sort_values(by="total_reviews", ascending= False)

df_sorted.reset_index(inplace=True)

df_top_100_sorted = df_sorted.head(n=100)

df_rest_sorted = df_sorted.iloc[100:20000]
```

```python
filepath_top100 = "top_100_games_total_reviews.json.gz"
filepath_rest = "rest_games_total_reviews.json.gz"

with gzip.open(filepath_top100,"wt", encoding="utf-8") as f:
        df_top_100_sorted.to_json(
        f,
        orient="records",
        force_ascii=False
    )
    
with gzip.open(filepath_rest,"wt", encoding="utf-8") as f:
        df_rest_sorted.to_json(
        f,
        orient="records",
        force_ascii=False
    )
```
